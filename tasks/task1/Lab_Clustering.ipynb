```python
class StandardScaler:
    def __init__(self):
        self.mean=None
        self.std =None
    def fit(self,X):
        self.mean= np.mean(X,axis=0)
        self.std= np.std(X,axis=0)
        return self
    def transform(self, X):
        return (X-self.mean)/self.std

    def fit_transform(self, X):
        return self.fit(X).transform(X)
        
```

```python
from numpy.random import uniform
import random


def most_common(lst):
    return max(set(lst), key=lst.count)


def euclidean(point, data):
    return np.sqrt(np.sum((point - data)**2, axis=1))


class KMeans:

    def __init__(self, n_clusters=8, max_iter=300):
        self.n_clusters = n_clusters
        self.max_iter = max_iter

    def fit(self, X_train):

       
        self.centroids = [random.choice(X_train)]

        for _ in range(self.n_clusters-1):
            dists = np.sum([euclidean(centroid, X_train) for centroid in self.centroids], axis=0)
            dists /= np.sum(dists) # растояние по нормализации  
            new_centroid_idx = np.random.choice(range(len(X_train)), size=1, p=dists)[0]  
            self.centroids += [X_train[new_centroid_idx]]

        iteration = 0
        while True:
            sorted_points = [[] for _ in range(self.n_clusters)]
            for x in X_train:
                dists = euclidean(x, self.centroids)
                centroid_idx = np.argmin(dists)
                sorted_points[centroid_idx].append(x)

            prev_centroids = self.centroids
            self.centroids = [np.mean(cluster, axis=0) for cluster in sorted_points]
            for i, centroid in enumerate(self.centroids):
                if np.isnan(centroid).any():  
                    self.centroids[i] = prev_centroids[i]
            iteration += 1
            if not(np.not_equal(self.centroids, prev_centroids).any() and iteration < self.max_iter):
                break

    def evaluate(self, X):
        centroids = []
        centroid_idxs = []
        for x in X:
            dists = euclidean(x, self.centroids)
            centroid_idx = np.argmin(dists)
            centroids.append(self.centroids[centroid_idx])
            centroid_idxs.append(centroid_idx)

        return centroids, centroid_idxs
```

```python
def calinski_harabasz_score(X, labels):
    n_samples, _ = X.shape
    n_labels = len(set(labels))


    extra_disp, intra_disp = 0.0, 0.0
    mean = np.mean(X, axis=0)
    for k in set(labels):
        cluster_k = X[labels == k]
        mean_k = np.mean(cluster_k, axis=0)
        extra_disp += len(cluster_k) * np.sum((mean_k - mean) ** 2)
        intra_disp += np.sum((cluster_k - mean_k) ** 2)

    return (
        1.0
        if intra_disp == 0.0
        else extra_disp * (n_samples - n_labels) / (intra_disp * (n_labels - 1.0))
    )
```

```python
from collections import defaultdict
class DBSCAN:
    def __init__(self, eps=1, min_samples=5):
        self.eps = eps
        self.min_samples = min_samples

    def fit(self, X):
        clusters = defaultdict(set)
        clusters_now = np.empty(len(X))
        clusters_now.fill(-1)
        new_cluster = 0
        def merge_clusters(left,rigth):
            rigth_cluster_key =clusters_now[rigth]
            clusters[clusters_now[left]]|=clusters[clusters_now[rigth]]
            for cls in clusters[clusters_now[rigth]]:
                clusters_now[cls] = clusters_now[left]
            clusters[rigth_cluster_key].clear() 

            
        for index,data_i in enumerate(X):
            for jndex, data_j in enumerate(X):
                if index == jndex:
                    continue
                
                dist = np.sqrt(np.sum((data_i - data_j)**2))
                # print(dist)
                if dist<self.eps:
                    if clusters_now[index]!= -1 or clusters_now[jndex]!=-1:
                        if clusters_now[index]!=-1 and clusters_now[jndex]!=-1:
                            merge_clusters(index,jndex)
                        else:
                            if clusters_now[index]!=-1:
                                clusters_now[jndex] = clusters_now[index]
                                clusters[clusters_now[index]].add(jndex)
                            else:
                                clusters_now[index] = clusters_now[jndex]
                                clusters[clusters_now[jndex]].add(index)
                                
                    else:
                        clusters_now[index] = new_cluster
                        clusters_now[jndex] = new_cluster
                        clusters[new_cluster]|={index,jndex}
                        new_cluster+=1



        for key, value in clusters.items():
            if len(value)< self.min_samples:
                for i in value:
                    clusters_now[i]=-1
        self.labels =clusters_now
        return clusters_now
```

```python
import optuna
optuna.logging.set_verbosity(optuna.logging.WARNING)
def objective(trial):
    params = {
        "eps":trial.suggest_float("eps", 1.0, 15.0),
        "min_samples": trial.suggest_int("min_samples",1,15)
    }
    if calinski_harabasz_score(data, DBSCAN(**params).fit(data)) == float('inf'):
        return 2
    return calinski_harabasz_score(data, DBSCAN(**params).fit(data))



study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=2000, timeout=1200*4, n_jobs=8)
print("Number of finished trials: {}".format(len(study.trials)))
trial = study.best_trial
print("  Value: {}".format(trial.value))
print(trial.params)
```
